```
#MIN = 800*(x1+x2)+900*(x3+x6)+1000*(x4+x5)
#sj a: 
#(turno 1) x1+x6>=51
#(turno 2) x1+x2>=58
#(turno 3) x2+x3>=62
#(turno 4) x3+x4>=41
#(turno 5) x4+x5>=32
#(turno 6) x5+x6>=19, 
#sendo x1 = trab que começam no primeiro turno, x2 = trab que começam no segundo turno, ...

# Definindo o problema como de maximização
prob = pulp.LpProblem('Exercício de Aula', pulp.LpMinimize)

# Definindo as variáveis de decisão
x1 = pulp.LpVariable('Trab que começam no primeiro turno', lowBound=0, cat='Integer')
x2 = pulp.LpVariable('Trab que começam no segundo turno', lowBound=0, cat='Integer')
x3 = pulp.LpVariable('Trab que começam no terceiro turno', lowBound=0, cat='Integer')
x4 = pulp.LpVariable('Trab que começam no quarto turno', lowBound=0, cat='Integer')
x5 = pulp.LpVariable('Trab que começam no quinto turno', lowBound=0, cat='Integer')
x6 = pulp.LpVariable('Trab que começam no sexto turno', lowBound=0, cat='Integer')

#Definindo a função a ser minimizada
MIN=800*(x1+x2)+900*(x3+x6)+1000*(x4+x5)

#Adicionando a função-objetivo
prob += MIN

#Definindo a função do primeiro turno
turno1 = x1 + x6

#Adicionando a função do primeiro turno nas restrições
prob += (turno1>=51)

#Definindo a função do segundo turno
turno2 = x1 + x2

#Adicionando a função do segundo turno nas restrições
prob += (turno2>=58)

#Definindo a função do terceiro turno
turno3 = x2 + x3

#Adicionando a função do terceiro turno nas restrições
prob += (turno3>=62)

#Definindo a função do quarto turno
turno4 = x3 + x4

#Adicionando a função do quarto turno nas restrições
prob += (turno4>=41)

#Definindo a função do quinto turno
turno5 = x4 + x5

#Adicionando a função do quinto turno nas restrições
prob += (turno5>=32)

#Definindo a função do sexto turno
turno6 = x5 + x6

#Adicionando a função do sexto turno nas restrições
prob += (turno6>=19)

#escrevendo o problema de otimização linear
print(prob)

# Resolvendo o problema 
optimization_result = prob.solve()

# Verificando se a solução ótima foi encontrada
assert optimization_result == pulp.LpStatusOptimal

#mostrando o resultado
for var in (x1, x2, x3, x4, x5, x6):
    print('Escala semanal ótima é de {}: {:1.0f}'.format(var.name, var.value()))
```
